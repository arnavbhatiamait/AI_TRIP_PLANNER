```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq
load_dotenv()
llm = ChatGroq(model="deepseek-r1-distill-llama-70b",api_key=os.getenv("GROQ_API_KEY"), temperature=0.1, max_tokens=1000)
llm.invoke("what is use of * in python")
```

```python
from langchain.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """   Multiplies two integers."""
    return a * b
```
